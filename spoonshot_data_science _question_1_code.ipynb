```python
munchboxfile=open("/content/munchbox.txt")
lines=munchboxfile.readlines()
print(lines)
#stripping of symbols
k=0
for i in lines:
    lines[k]=i.strip("',\n[]’?‘-@")
    k+=1
print(lines)
#tokenizing
article = []
for s in lines:
    article.append(word_tokenize(s.lower()))
print(article)
```

```python
ingredients_list=open("/content/ingredients.txt")
queries=ingredients_list.readlines()
print(queries)
#queries stripping
j=0
for i in queries:
    queries[j]=i.strip("',\n[]’‘")
    j+=1
print(queries)
#queries tokenizing
queries_token=[]
for s in queries:
    queries_token.append(word_tokenize(s.lower()))
print(queries_token)
```

```python
def cosine(u, v):
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
matrix = []
for r in range(0, len(queries)):
  matrix.append([cosine(query_encoding[r], sentence_embeddings[c]) for c in range(0, 60)])
print(matrix)
```

```python
scores_sum=[]
for r in range(0, len(queries)):
  temp=0
  for c in range(0, len(article)):
    temp+=matrix[r][c]
  scores_sum.append(temp)
```

```python
print(scores_sum)
d={'scores':scores_sum, 'query':queries}
df=pd.DataFrame(d)
```
